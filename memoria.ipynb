Se pasa el  dataset a español renombrando las columnas de la siguiente manera:

> - age  &rarr;  edad
> - gender  &rarr;  genero
> - daily_social_median_hours  &rarr;  horas_diarias_redes
> - platform_usage  &rarr;  plataforma_usada
> - sleep_hours  &rarr; horas_sueño
> - screen_time_before_sleep  &rarr;  tiempo_pantalla_antes_dormir
> - academic_performance  &rarr;  rendimiento_academico
> - physical_activity  &rarr;  actividad_fisica
> - social_interaction_level  &rarr;  nivel_interacción_social
> - sttress_level  &rarr;  nivel_estres
> - anxiety_level &rarr;  nivel_ansiedad
> - addiction_level  &rarr;  nivel_adiccion
> - depression_label  &rarr;  depresion


Y los valores del dataset que son en inglés que corresponden a las siguientes columnas:
- gender  &rarr;  genero

    Se modifica:
    * male  &rarr;  masculino
    * female  &rarr;  femenino

- social_interaction_level  &rarr;  nivel_interacción_social

    Se modifica:
    * low  &rarr;  baja
    * medium  &rarr; media
    * high  &rarr;  alta 





```python
# Renombrar columnas en español
datos_e = datos.rename(columns={
    'age': 'edad', 
    'gender': 'genero',
    'daily_social_media_hours': 'horas_diarias_redes',
    'platform_usage': 'plataforma_usada',
    'sleep_hours': 'horas_sueño',
    'screen_time_before_sleep': 'tiempo_pantalla_antes_dormir',
    'academic_performance': 'rendimiento_academico',
    'physical_activity': 'actividad_fisica',
    'social_interaction_level': 'nivel_interaccion_social',
    'stress_level': 'nivel_estres',
    'anxiety_level': 'nivel_ansiedad',
    'addiction_level': 'nivel_adiccion',
    'depression_label': 'depresion'})

# Renombramos los datos de la columna genero
datos_e['genero'] = datos_e['genero'].replace('male', 'masculino')
datos_e['genero'] = datos_e['genero'].replace('female', 'femenino')

# Renombramos los datos de la columna nivel_interaccion_social
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('low', 'baja')
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('medium', 'media')
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('high', 'alta')
datos_e

# datos --> dataset inicial
# datos_e --> dataset renombrado
```
