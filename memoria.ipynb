## 3. MODIFICACIONES EN LOS DATOS

Se pasa el  dataset a español renombrando las columnas de la siguiente manera:

> - age  &rarr;  edad
> - gender  &rarr;  genero
> - daily_social_median_hours  &rarr;  horas_diarias_redes
> - platform_usage  &rarr;  plataforma_usada
> - sleep_hours  &rarr; horas_sueño
> - screen_time_before_sleep  &rarr;  tiempo_pantalla_antes_dormir
> - academic_performance  &rarr;  rendimiento_academico
> - physical_activity  &rarr;  actividad_fisica
> - social_interaction_level  &rarr;  nivel_interacción_social
> - sttress_level  &rarr;  nivel_estres
> - anxiety_level &rarr;  nivel_ansiedad
> - addiction_level  &rarr;  nivel_adiccion
> - depression_label  &rarr;  depresion


También se pasa a español los valores del dataset que son en inglés que corresponden a las siguientes columnas:
- gender  &rarr;  genero

    Se modifica:
    * male  &rarr;  masculino
    * female  &rarr;  femenino

- social_interaction_level  &rarr;  nivel_interacción_social

    Se modifica:
    * low  &rarr;  baja
    * medium  &rarr; media
    * high  &rarr;  alta 





```python
# Renombrar columnas en español
datos_e = datos.rename(columns={
    'age': 'edad', 
    'gender': 'genero',
    'daily_social_media_hours': 'horas_diarias_redes',
    'platform_usage': 'plataforma_usada',
    'sleep_hours': 'horas_sueño',
    'screen_time_before_sleep': 'tiempo_pantalla_antes_dormir',
    'academic_performance': 'rendimiento_academico',
    'physical_activity': 'actividad_fisica',
    'social_interaction_level': 'nivel_interaccion_social',
    'stress_level': 'nivel_estres',
    'anxiety_level': 'nivel_ansiedad',
    'addiction_level': 'nivel_adiccion',
    'depression_label': 'depresion'})

# Renombramos los datos de la columna genero
datos_e['genero'] = datos_e['genero'].replace('male', 'masculino')
datos_e['genero'] = datos_e['genero'].replace('female', 'femenino')

# Renombramos dentro de la columna plataforma_usada --> both a ambas
datos_e['plataforma_usada'] = datos_e['plataforma_usada'].replace('Both', 'Ambas')

# Renombramos los datos de la columna nivel_interaccion_social
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('low', 'baja')
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('medium', 'media')
datos_e['nivel_interaccion_social'] = datos_e['nivel_interaccion_social'].replace('high', 'alta')
datos_e

# datos --> dataset inicial
# datos_e --> dataset renombrado
```

## 5. HIPOTESIS

### 5.1  A mayor tiempo de exposición a redes sociales antes de dormir menor es el número de horas de sueño.


```python
# Crea el gráfico
sns.regplot(
    data=datos_e,
    x="tiempo_pantalla_antes_dormir",
    y="horas_sueño",
    scatter_kws={
        "alpha": 0.6,
        "color": "#2c3e50",
    },  # Puntos semi-transparentes oscuros
    line_kws={
        "color": "#e74c3c",
        "linewidth": 2.5,
    },  # Línea de tendencia roja
)

# etiquetas y título
plt.title(
    "Impacto del tiempo de pantalla previo al sueño en el descanso",
    fontsize=14,
    pad=15,
)
plt.xlabel("Tiempo de pantallas antes de dormir (Horas)", fontsize=12)
plt.ylabel("Horas de sueño", fontsize=12)
plt.savefig('img/impacto_tiempo_pantalla.png');

```

La línea roja es prácticamente horizontal y los puntos están dispersos por todo el lienzo, el dataset indica que no existe una relación lineal entre el tiempo de pantalla antes de dormir y las horas de sueño.


```python
correlacion = datos_e["tiempo_pantalla_antes_dormir"].corr(datos_e["horas_sueño"])
print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}")
```

El valor de 0.0102 confirma matemáticamente que la relación lineal es  irrelevante.

#### CONCLUSION:

La hipótesis no se cumple en este conjunto de datos.
