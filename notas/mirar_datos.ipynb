```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('datos_sueño_movil.csv')

# Visualizar las primeras filas para asegurar que se cargó bien
print(df.head())
# Comprobar tipos de datos y valores nulos
print(df.info())
Usa el código con precaución.2. Limpieza y Preparación (EDA)Asegúrate de que las columnas de "horas de móvil" y "horas de sueño" sean numéricas. Elimina filas con valores vacíos (NaN) si es necesario.python# Eliminar nulos si los hay en las columnas clave
df = df.dropna(subset=['horas_movil', 'horas_sueño'])

# Si las columnas no son numéricas, convertirlas
# df['horas_movil'] = pd.to_numeric(df['horas_movil'])
Usa el código con precaución.3. Análisis de CorrelaciónEl método .corr() calcula la correlación de Pearson, que mide la fuerza de la relación lineal entre dos variables.python# Calcular la matriz de correlación
correlacion = df[['horas_movil', 'horas_sueño']].corr()
print(correlacion)
Usa el código con precaución.Interpretación:Cerca de -1: Fuerte correlación negativa (A más móvil, mucho menos sueño).Cerca de 0: Sin relación aparente.Cerca de 1: Fuerte correlación positiva (A más móvil, más sueño - poco probable).4. Visualización de los DatosPara ver claramente la tendencia, lo mejor es un diagrama de dispersión (scatter plot).pythonplt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='horas_movil', y='horas_sueño')
plt.title('Relación: Horas de Móvil vs Horas de Sueño')
plt.xlabel('Horas de móvil antes de dormir')
plt.ylabel('Horas de sueño')
plt.grid(True)
plt.show()
Usa el código con precaución.5. Análisis Avanzado (Agrupación)Puedes agrupar por horas de móvil para ver el promedio de sueño.python# Agrupar y calcular promedio de sueño por horas de móvil
estudio_agrupado = df.groupby('horas_movil')['horas_sueño'].mean().reset_index()
print(estudio_agrupado)
```
